## 1. Importação de bibliotecas e carregamentos dos dados.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Estilo dos gráficos
sns.set(style="whitegrid", palette="muted")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# Lendo arquivo
df = pd.read_csv('../docs/calca_jeans.csv', sep=';')
```

## 2. Análise e tratamento dos dados

```python
# Visão geral do dataset
print("Dimensões do dataset:", df.shape)
display(df.head())
display(df.info())
display(df.describe(include='all'))
```

```python
# Verificando valores nulos
print("\n Valores nulos por coluna:")
missing = df.isnull().sum()     # Salva a soma dos valores nulos em uma variável
print("\n Resultado: ", missing[missing > 0])     # Exibe apenas se a soma for maior que 0

# Exibe de forma mais 'vertical' os dados
print("\n Exibindo dados de forma vertical:")
print(df.isnull().sum().to_frame('Missing_Values').T)
```

```python
# Conversão de colunas numéricas que estão como object

# Lista de colunas que devem ser numéricas mas estão como object
cols_to_convert = ['PrecoVenda', 'PrecoCusto', 'TotalDesconto', 'Frete', 'CustoMaoDeObra']

# Substituir vírgulas por ponto e converter para float
for col in cols_to_convert:
    df[col] = df[col].str.replace(',', '.').astype(float)

# Conferir os tipos após a conversão
display(df[cols_to_convert].dtypes)
```

```python
# Convertendo object para datetime
df['DataNotaFiscal'] = pd.to_datetime(df['DataNotaFiscal'], dayfirst=True)
```

```python
# Criando uma coluna de notas com valores tipo float baseado na satisfacao dos clientes:
df['notas'] = df['Indi_Satisfacao'].str.extract(r'(\d+(?:\.\d+)?)').astype(float)
print("Coluna de notas: ")
print(df['notas'])
```

```python
# Distribuição das notas:
sns.histplot(df['notas'], bins=10, kde=True)
plt.title('Distribuição das Notas de Satisfação')
plt.xlabel('Nota')
plt.ylabel('Frequência')
plt.show()
```

```python
sns.boxplot(data=df, x='notas')  # boxplot de notas
```

### Análise categórica dos dados

```python
# Análise das variáveis categóricas:
cat_cols = df.select_dtypes(include='object').columns.drop('Indi_Satisfacao')

for col in cat_cols:
    plt.figure(figsize=(10, 5))  # Tamanho ajustável
    ax = sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette='pastel')
    plt.title(f'Distribuição de {col}', fontsize=14, weight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Adiciona os rótulos de valor nas barras
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f'{height}', 
                    (p.get_x() + p.get_width() / 2., height), 
                    ha='center', va='bottom', fontsize=9, color='black', rotation=0)

    plt.tight_layout()
    plt.show()
```

### Correlações

```python
# Correlação entre variáveis numéricas:

num_cols = df.select_dtypes(include=np.number)

correlation = num_cols.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".3f", linewidths=1)
plt.title("Mapa de Correlação entre Variáveis Numéricas")
plt.show()
```
